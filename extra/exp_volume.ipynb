```python
import numpy as np
import matplotlib.pyplot as plt

def plot(volume, title='Histogram of Volume Values'):
    flattened_values = volume.flatten()

    plt.figure(figsize=(6, 3))
    plt.hist(flattened_values, bins=50, color='blue', alpha=0.7, edgecolor='black')
    plt.xlabel("Intensity Value")
    plt.ylabel("Frequency")
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
volume_gt = np.load("../data/z-bin/battery/method_1/image_gt.npy")
plot(volume_gt, title='GT')
```

```python
volume_pred = np.load("../data/z-bin/battery/method_3/image_pred.npy")
plot(volume_pred, title='raw')
```

```python
volume_pred_exp = np.load("../data/z-bin/battery/method_3/image_pred_exp.npy")
volume_pred_exp = (volume_pred_exp - np.min(volume_pred_exp)) / (np.max(volume_pred_exp) - np.min(volume_pred_exp))
plot(volume_pred_exp, title='exp(-pred)')
np.save('../data/z-bin/battery/method_3/image_pred_exp_norm.npy', volume_pred_exp)
```
